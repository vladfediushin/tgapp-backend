```python
import pandas as pd
import json
```

```python
input_path = '/Users/vladfediushin/MEGA/coding/tgapp_driving_final_v3/data/1/questions.csv'
output_path = '/Users/vladfediushin/MEGA/coding/tgapp_driving_final_v3/questions.csv'
```

```python
df = pd.read_csv(input_path)
df.head()
```

```python
new_prefix = 'https://cdn.jsdelivr.net/gh/vladfediushin/driving-assets@main/Armenia/1/images/'

# Обработка колонки 'options'
def replace_in_options(options_str):
    try:
        options = json.loads(options_str)
        updated = [opt.replace("URLLINK/", new_prefix) if isinstance(opt, str) else opt for opt in options]
        return json.dumps(updated, ensure_ascii=False)
    except Exception:
        return options_str  # если не JSON — возвращаем как есть

df['options'] = df['options'].apply(replace_in_options)
df['question_image'] = df['question_image'].str.replace(r'^URLLINK/', new_prefix, regex=True)

df.head()
```

```python
# Функция для преобразования строки в JSONB-объект

def create_data_json(row):
    return {
        "question": row["question_itself"],
        "question_image": row["question_image"] if pd.notna(row["question_image"]) else None,
        "options": json.loads(row["options"]) if isinstance(row["options"], str) else [],
        "correct_index": int(row["correct_index"])
    }

# Применяем функцию к каждой строке
df["data"] = df.apply(create_data_json, axis=1)
```

```python
# Добавляем постоянные поля
df["topic"] = "Группа № 1. Маневрирование, построение на проезжей части"
df["country"] = "AM"
df["language"] = "ru"

# Оставляем только нужные поля и сохраняем
result_df = df[["data", "topic", "country", "language"]]

# Преобразуем столбец data в строку JSON
result_df["data"] = result_df["data"].apply(lambda x: json.dumps(x, ensure_ascii=False))

# Сохраняем
result_df.to_csv(output_path, index=False)
print(f"Saved to {output_path}")
```
